```python
fields = [('detailed_description', data.Field(sequential=True, tokenize=tokenizer, lower=True)),
          ('short_description', data.Field(sequential=True, tokenize=tokenizer, lower=True)),
          ('categories', data.Field(sequential=True, tokenize=tokenizer, lower=True)),
          ('genres', data.Field(sequential=True, tokenize=tokenizer, lower=True)),]

train_data = data.TabularDataset(path=game_file, format='csv', fields=fields, skip_header=True)
```

```python
# Define the path to the extracted GloVe embeddings file
glove_path = '/content/GloVe-1.2.zip'

# Define the TEXT field
TEXT = data.Field(sequential=True, tokenize=tokenizer, lower=True)

# Load the GloVe embeddings into the vocabulary
vectors = vocab.GloVe(name='6B', dim=100, cache='/content/glove/')  # Specify the dimensions (dim) as 100

# Build the vocabulary with the GloVe embeddings
TEXT.build_vocab(train_data, vectors=vectors)
```

```python
import torch.nn.utils.rnn as rnn_utils
```

```python
# Text preprocessing and embedding function
def text_to_tensor(detailed_description, short_description):
    detailed_tokens = TEXT.preprocess(detailed_description)
    short_tokens = TEXT.preprocess(short_description)

    # Ensure both sequences have the same length by padding the shorter one
    max_length = max(len(detailed_tokens), len(short_tokens))
    detailed_padded = [TEXT.vocab.stoi[token] for token in detailed_tokens] + [TEXT.vocab.stoi[TEXT.pad_token]] * (max_length - len(detailed_tokens))
    short_padded = [TEXT.vocab.stoi[token] for token in short_tokens] + [TEXT.vocab.stoi[TEXT.pad_token]] * (max_length - len(short_tokens))

    detailed_embed = torch.tensor(detailed_padded, dtype=torch.long).unsqueeze(0)  # Add batch dimension and set dtype
    short_embed = torch.tensor(short_padded, dtype=torch.long).unsqueeze(0)  # Add batch dimension and set dtype

    return torch.cat((detailed_embed, short_embed), dim=1)  # Concatenate along the second dimension

```

### Actual text processing on games

```python
EMBEDDING_DIM = vectors.vectors.shape[1]
```

```python
embeddings_list = []
max_length = 0  # Initialize max_length to zero

# Step 2: Process each row and obtain text embeddings
for index, row in text_data.iterrows():
    detailed_description = row['detailed_description']
    short_description = row['short_description']

    # Handle missing values by filling them with empty strings ('')
    detailed_description = str(detailed_description) if not pd.isnull(detailed_description) else ''
    short_description = str(short_description) if not pd.isnull(short_description) else ''

    embeddings = text_to_tensor(detailed_description, short_description)
    embeddings_list.append(embeddings)

    # Update the maximum length
    max_length = max(max_length, embeddings.shape[1])
```

```python
from torch.nn.utils.rnn import pad_sequence
```

```python
# Calculate the correct max_length based on the actual length of the longest sequence
max_length = max(embeddings_list[i].shape[1] for i in range(len(embeddings_list)))

# Pad all sequences to match the maximum length
padded_embeddings_list = [torch.nn.functional.pad(tensor, (0, max_length - tensor.shape[1], 0, 0), value=0) for tensor in embeddings_list]

# Stack all text embeddings into a single tensor
combined_text_embeddings = torch.stack(padded_embeddings_list, dim=0)
```

```python
from sklearn.cluster import KMeans

# Flatten the embeddings to 2D for clustering
num_samples, max_length, embedding_dim = combined_text_embeddings.size()
combined_text_embeddings = combined_text_embeddings.view(num_samples, -1)

# Number of clusters (you can adjust this based on your requirements)
num_clusters = 10

# Create the KMeans clustering model
kmeans_model = KMeans(n_clusters=num_clusters, random_state=42)

# Fit the model to the combined text embeddings
kmeans_model.fit(combined_text_embeddings)

# Get the cluster assignments for each game
cluster_assignments = kmeans_model.labels_

# Add the cluster assignments back to the original dataframe
text_data['cluster'] = cluster_assignments
```

```python
from sklearn.manifold import TSNE
```

```python
# Apply t-SNE for dimensionality reduction
# tsne = TSNE(n_components=2, random_state=42)
tsne = TSNE(n_components=2)

tsne_result = tsne.fit_transform(combined_text_embeddings)
```

```python
# Add the t-SNE embeddings to the DataFrame
text_data['x'] = tsne_result[:, 0]
text_data['y'] = tsne_result[:, 1]
```

```python
import plotly.graph_objects as go

# Create a scatter plot to visualize the clusters with images
colors = ['blue', 'green', 'red', 'purple', 'orange', 'yellow', 'cyan', 'magenta', 'lime', 'pink', 'teal', 'brown']

fig = go.Figure()

for cluster in range(num_clusters):
    cluster_df = text_data[text_data['cluster'] == cluster]
    name_list = cluster_df['name'].tolist()

    # Add traces for each cluster
    scatter_trace = go.Scatter(
        x=tsne_result[cluster_df.index, 0],
        y=tsne_result[cluster_df.index, 1],
        mode='markers',
        marker=dict(
            size=10,  # Set the size of the markers
            color=colors[cluster],  # Set the color based on the cluster
        ),
        # Use the name_list as the text to display on hover
        text=name_list,
        showlegend=False,
        # Assign the image URLs as customdata
        customdata=header_images,
        hovertemplate='<img src="%{customdata}" width="100px"><br>Game Name: %{text}<extra></extra>',
    )

    # Add the trace to the figure
    fig.add_trace(scatter_trace)

# Update the layout
fig.update_layout(
    title='Cluster Visualization with Images',
    xaxis=dict(title='X'),
    yaxis=dict(title='Y'),
    hovermode='closest',  # Show the closest data point's name on hover
    hoverlabel=dict(
        bgcolor="white",  # Set the background color of the hover labels
        font_size=12,    # Set the font size of the hover labels
    ),
)

# Show the plot
fig.show()

```
